```python
import math
import random
import matplotlib.pyplot as plt


def generate_Points(total_points):
    n1 = list()
    for i in range(total_points):
        x = random.random()
        y = random.random()
        n1.append((x,y))
    red_points, blue_points = get_Euclidean(n1)
    pi = (len(blue_points)/total_points)*4
    error = get_Error(pi)
    if(total_points == 10000):
        show_Plot(total_points,blue_points,red_points,n1)
    return pi, error
    

def get_Euclidean(n1):
    red = list()
    blue = list()
    for i in n1:
        squared_sum = i[0]**2 + i[1]**2
        if(math.sqrt(squared_sum) >= 1):
            red.append((i[0],i[1]))
        else:
            blue.append((i[0],i[1]))
    return red,blue


def get_Error(pi):
    return ((abs(pi-math.pi))/((pi+math.pi)/2))*100
        

def show_Plot(total_points, blue_points, red_points, n1):
    x = [i[0] for i in blue_points]
    y = [i[1] for i in blue_points]
    plt.scatter(x,y, color='b', s = .1)
    x = [i[0] for i in red_points]
    y = [i[1] for i in red_points]
    plt.scatter(x,y, color='r',s = .1)
    plt.gca().set_aspect('equal',adjustable = 'box')
    plt.xscale('linear')
    plt.yscale('linear')
    plt.plot()
    
def main():
    pi, error = generate_Points(1000)
    print("n = 10 ^ 3 pi = %.6f error = %.6f%s" % (pi,error,"%"))
    
    pi, error = generate_Points(10000)
    print("n = 10 ^ 4 pi = %.6f error = %.6f%s" % (pi,error,"%"))
    
    pi, error = generate_Points(100000)
    print("n = 10 ^ 5 pi = %.6f error = %.6f%s" % (pi,error,"%"))
    
    pi, error = generate_Points(1000000)
    print("n = 10 ^ 6 pi = %.6f error = %.6f%s" % (pi,error,"%"))

    
main()
    
        
    
```
