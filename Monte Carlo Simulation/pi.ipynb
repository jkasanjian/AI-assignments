```python
import math
import random
import matplotlib.pyplot as plt


def generate_points(n):
    points = []
    for i in range(n):
        points.append((random.random(), random.random()))
    return points
                      
                      
def calculate(points):
    blue = []
    red = []
    
    for p in points:
        if distance_origin(p) < 1:
            blue.append(p)
        else:
            red.append(p)

    pi = 4 * (len(blue)/len(points))
    
    return pi, red, blue
              
    
def distance_origin(p):
    return math.sqrt((p[0]**2) + (p[1]**2))


def calculate_error(pi):
    return ( (abs((pi - math.pi)))/((pi+math.pi)/2) * 100)


def show_graph(graph):
    red, blue = graph[0], graph[1]
    x = list((p[0] for p in red))
    y = list((p[1] for p in red))
    plt.scatter(x, y, color='r', s=.1)
    x = list((p[0] for p in blue))
    y = list((p[1] for p in blue))
    plt.scatter(x, y, color='b', s=.1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Estimating pi using simulation with n = 10 ^ 4')
    plt.show()
    


def main():
    N = [10**3, 10**4, 10**5, 10**6]
    pi_list = []
    graph = []
    for i in range(len(N)):
        pts = generate_points(N[i])
        pi, red, blue = calculate(pts)
        pi_list.append(pi)
        if i == 1:
            graph.append(red)
            graph.append(blue)
    
    error_list = []
    for pi in pi_list:
        error_list.append(calculate_error(pi))
        
    for i in range(len(N)):
        print('n = 10 ^ %d pi = %.6f  error =  %.4f %s'%(i+3, pi_list[i], error_list[i], '%'))
        
    show_graph(graph)
    
    
    

main()
    
```
